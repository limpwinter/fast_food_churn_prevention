```python
from sklearn.ensemble import StackingClassifier
# from models import CatBoostModel
from src import MetaInfo
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import pandas as pd
```

```python
train_data = pd.read_parquet("data/train.parquet")
test_data = pd.read_parquet("data/test.parquet")
```

```python
train_mng = MetaInfo(train_data)
test_mng = MetaInfo(test_data)
```

```python
X_train = train_mng.generate_meta_features()
y_train = train_mng.get_labels()
```

```python
X_test = test_mng.generate_meta_features()
```

```python


catboost_clf = CatBoostClassifier()

xgb_clf = XGBClassifier(
   # n_estimators=2000,
   # max_depth=4,
   # learning_rate=0.06,

)

random_forest_clf = RandomForestClassifier()

estimators = [
    ('catboost', catboost_clf),
    ('xgb', xgb_clf),
    ('random_forest', random_forest_clf)
]

clf = StackingClassifier(
    estimators=estimators,
    final_estimator=CatBoostClassifier(),
)
```

```python
clf.fit(X_train.values,y_train['buy_post'].values)
```
